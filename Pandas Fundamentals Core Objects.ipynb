# 4️. dtype (data type)

## It tells what kind of data is in column

```python
df = pd.DataFrame({
    "Age": [20, 21, 22],
    "Name": ["A", "B", "C"]
})
```

```python
df.dtypes
```

# type conversion

```python
df["Age"]=df["Age"].astype(float)
```

```python
df.dtypes
```
