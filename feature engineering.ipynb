```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "John", "Riya", "Aman"],
    "Age": [22, 25, 23, 30, 29],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai"],
    "Salary": [50000, 60000, 52000, 80000, 75000],
    "Experience": [1, 3, 2, 7, 6]
}


```

```python
df = pd.DataFrame(data)
```

```python
df
```

## example: salary in lakhs

```python
df["Salary_lakhs"]=df["Salary"]/100000
```

```python
df
```

## example: age after 5 years

```python
df["Age_after_5yrs"]=df["Age"]+5
```

```python
df
```

# 2. mathematical operations

## Apply arithmetic operations directly on columns

##  Example: Salary per year of experience

```python
df["Salary_per_exp"]=df["Salary"]/df["Experience"]
```

```python
df
```

# 3. Conditional Columns

## Create binary / category features based on conditions.

## Two main ways:

## np.where() (FAST)
## apply() (flexible but slower)

## using np.where()

```python
df["High_Salary"]=np.where(df["Salary"]>60000,1,0)
```

```python
df
```

## Using apply()


```python
df["Senior_level"]=df["Experience"].apply(lambda x: "Senior" if x >=5 else "junior")
                                         
```

```python
df
```

# 4. String Operations (str accessor)

## Used when columns contain text

## Lowercase names

```python
df["Name_lower"] = df["Name"].str.lower()
```

```python
df
```

## City starts with D?

```python
df["Starts_D"]=df["City"].str.startswith("D")
```

```python
df
```

## Replace text

```python
df["City"]=df["City"].str.replace("Mumbai","Bombay")
```

```python
df
```

## Categorical Encoding

```python
import pandas as pd

df1 = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Pune", "Delhi"],
    "Salary": [50000, 60000, 55000, 52000]
})

encoded = pd.get_dummies(df1, columns=["City"])
print(encoded)
```
